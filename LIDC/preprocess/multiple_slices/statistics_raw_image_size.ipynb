```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
base = '/data/liumingzhou/CounterAlign_output/preprocess/multiple_slices/mindiam7_fulllung_cls'
```

```python
sizes = list()
for file in os.listdir(os.path.join(base,'train')):
    img = np.load(os.path.join(base,'train',file))
    sizes.append(list(img.shape))
sizes = np.array(sizes)
```

```python
plt.scatter(sizes[:,0],sizes[:,1])
plt.title('raw image size distribution: {}'.format(base.split('/')[-1]))
plt.xlabel('w')
plt.ylabel('h')
```

```python
fig,ax = plt.subplots(1,5,figsize=(15,3))
for ind in range(5):
    file = os.listdir(os.path.join(base,'train'))[ind]
    img = np.load(os.path.join(base,'train',file))
    ax[ind].imshow(img,cmap='gray')
    ax[ind].axis('off')
```
